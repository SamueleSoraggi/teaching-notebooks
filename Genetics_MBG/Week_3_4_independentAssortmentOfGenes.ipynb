```python
import random
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# --- Step 1: Define alleles for the two traits ---
eye_alleles = ["B", "b"]   # Brown / blue
hair_alleles = ["H", "h"]  # Black / blond
```

```python
def make_gamete():
    """Randomly choose one allele for eyes and one for hair (independent)."""
    return random.choice(eye_alleles) + random.choice(hair_alleles)

def make_offspring():
    """Combine gametes from two parents into one offspring genotype."""
    g1 = make_gamete()
    g2 = make_gamete()
    return g1 + g2   # string with 4 letters (2 eye alleles + 2 hair alleles)
```

```python
# --- Step 2: Simulate many offspring ---
N = 100000  # number of offspring
offspring = [make_offspring() for _ in range(N)]
```

```python
# --- Step 3: Convert genotype -> phenotype ---
def phenotype(genotype):
    # First and third letters are eye alleles
    eyes = genotype[0] + genotype[2]
    # Second and fourth letters are hair alleles
    hair = genotype[1] + genotype[3]

    eye_trait = "Brown eyes" if "B" in eyes else "Blue eyes"
    hair_trait = "Black hair" if "H" in hair else "Blond hair"
    return f"{eye_trait}, {hair_trait}"
```

```python
phenotypes = [phenotype(child) for child in offspring]
```

```python
# --- Step 4: Count how often each phenotype appears ---
counts = Counter(phenotypes)

# --- Step 5: Plot results ---
plt.bar(counts.keys(), counts.values())
plt.xticks(rotation=30, ha="right")
plt.ylabel("Number of offspring")
plt.title("Independent Assortment Simulation (BbHh x BbHh)")
plt.show()
```
